### Some visualizations

```python
attr_list = ['revenue', 'budget', 'popularity', 'vote_avg']
title_list = ['Filmes com maior receita', 
              'Filmes com maior despesa', 
              'Filmes mais populares', 
              'Filmes mais bem avaliados']
i=0
k=10  #top-k for an attr
for attr in attr_list:
    movies_attr_dict = utils.sort_movies_by_attr(movies_dict, attr, True)
    
    movies = list(movies_attr_dict.keys())[:k]
    attrs  = list(movies_attr_dict.values())[:k]
    plt.figure(figsize=(10,8))
    plt.barh(movies, attrs, 0.7, align='center', color="lightskyblue")
    plt.xticks(fontsize=12)
    plt.ylabel(attr, fontsize=14)
    plt.yticks(range(len(attrs)), ["{:.2f}".format(v) for v in attrs], fontsize=12)
    for index, label in enumerate(movies):
        plt.text(1, index, label, ha='left',va='center')
    plt.title(title_list[i], fontsize=16)
    plt.gca().invert_yaxis()
    plt.show()
    i+=1
```

```python
k = 10
genres = list(genres2plot.keys())[:k]
count  = list(genres2plot.values())[:k]

plt.figure(figsize=(16,8))
plt.bar(genres, count, 0.7, align='center', color="lightskyblue")
plt.xticks(rotation=70, fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('Quantidade de filmes', fontsize=14)
plt.xlabel('Gênero', fontsize=14)
for i,v in enumerate(count):
    plt.text(x=i-0.2, y=v+100, s=f"{v}" , fontdict=dict(fontsize=12))
plt.title('Quantidade de filmes por gênero', fontsize=16)
plt.show()
```
