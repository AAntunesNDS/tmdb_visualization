## Utils functions

```python
def get_genres_from_row(row):
    aux_row = row[1:-1]
    raw_genres = aux_row.split('}')
    genres = list()
    for g in raw_genres:
        aux = g.split(':')
        if len(aux) >= 3:
            genre = aux[2].replace("'", "")
            genres.append(genre)
    return genres
```

```python
def build_movie_attr_dict(movies_dict, attr):
    attr_dict = dict()
    for _id, movie_dict in movies_dict.items():
#         movie_id = f"{_id}_{movie_dict['title']}"
        movie_id = f"{movie_dict['title']}"
        attr_dict[movie_id] = movie_dict[attr]
    return attr_dict
```

```python
def sort_movies_by_attr(movies_dict, attr, desc=True):
    attr_dict = build_movie_attr_dict(movies_dict, attr)
    return {k: v for k, v in sorted(attr_dict.items(), key=lambda item: item[1], reverse=desc)}
```

```python
aux_df = df_movies_meta.head(18000)
# aux_df = df_movies_meta.copy()
movies_dict = dict()

# remove this gambiarra after


for i, row in aux_df.iterrows():
    id_movie = row['id']
    inner_dict = dict(
        title=row['original_title'],
        budget=float(row['budget']),
        vote_count=float(row['vote_count']),
        vote_avg=float(row['vote_average']),
        revenue=float(row['revenue']),
        genres=get_genres_from_row(row['genres']),
        runtime=float(row['runtime']),
        popularity=float(row['popularity']),
        profit=float(row['budget'])-float(row['revenue'])
    )   
    movies_dict[id_movie] = inner_dict
display(movies_dict)
```

### Some visualizations

```python
attr_list = ['revenue', 'budget', 'popularity', 'vote_avg']
title_list = ['Filmes com maior receita', 
              'Filmes com maior despesa', 
              'Filmes mais populares', 
              'Filmes mais bem avaliados']
i=0
k=10  #top-k for an attr
for attr in attr_list:
    movies_attr_dict = sort_movies_by_attr(movies_dict, attr, True)
    
    movies = list(movies_attr_dict.keys())[:k]
    attrs  = list(movies_attr_dict.values())[:k]
    plt.figure(figsize=(10,8))
    plt.barh(movies, attrs, 0.7, align='center', color="lightskyblue")
    plt.xticks(fontsize=12)
    plt.ylabel(attr, fontsize=14)
    plt.yticks(range(len(attrs)), ["{:.2f}".format(v) for v in attrs], fontsize=12)
    for index, label in enumerate(movies):
        plt.text(1, index, label, ha='left',va='center')
    plt.title(title_list[i], fontsize=16)
    plt.gca().invert_yaxis()
    plt.show()
    i+=1
```

```python
%%time
genres_dict = build_movie_attr_dict(movies_dict, 'genres')


genres2plot = dict()
for movie, genres in genres_dict.items():
    for genre in genres:
        genre = genre[1:]
        if genre in genres2plot:
            genres2plot[genre] += 1
        else:
            genres2plot[genre] = 1
```
